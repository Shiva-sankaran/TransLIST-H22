```python
def return_max_len(lines):
    lengths = []
    curr_len = 0
    for i in range(len(lines)):
        curr_len+=1
        if(lines[i] == '\n'):
            lengths.append(curr_len)
            curr_len = 0
    return max(lengths)            
            
            
        
    
```

```python
return_max_len(lines=lines)
```

```python
# remove overflow
curr_len = 0
new_lines = []
noverflow = 0
do_not_add = False
curr_lines = []
skip = False
for i in range(len(lines)):
    if(skip):
        if(lines[i] == "\n"):
            skip = False
            curr_lines = []
            curr_len = 0
            
        continue

    curr_len+=1
    curr_lines.append(lines[i])

    if(lines[i] == '\n'):
        new_lines.extend(curr_lines)
        curr_lines = []
        curr_len = 0
        continue
        
    if(curr_len>150):
        skip = True
        noverflow+=1
        continue

              
```
